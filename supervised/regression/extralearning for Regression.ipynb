# extralearning for Regression
----

Regression is a type of statistical modeling technique used in machine learning and statistics. Its primary purpose is to establish a relationship between a dependent variable (or target) and one or more independent variables (or predictors). This relationship is often expressed in the form of an equation that can be used to make predictions.

Read more: https://en.wikipedia.org/wiki/Regression_analysis

```python
#!pip install extralearning

# Regression is containg within extralearning.supervised.Regression class

from extralearning.supervised import Regression
from sklearn.datasets import load_diabetes

# Constant random state
RANDOM_STATE = 102
```

```python
# Loading sklearn diabetes data

X, y = load_diabetes(return_X_y = True,
                     as_frame = True)
```

```python
# Creating the extralearning regressor

model = Regression(n_jobs = -1,
                   ignore_warnings = True)
```

```python
# Fit and train the classifier

model.fit_train(X, y, CV = 2, CV_params = None, verbose = True)
```

```python
# Creating summary

summary = model.summary(pandas = True)

summary
```
